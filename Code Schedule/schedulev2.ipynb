````python
import random

# Data jadwal dan matakuliah
#jadwal = {
#    "MII225201": {"matakuliah": "Analisis Algoritme", "hari": 2, "jam": 1, "ruangan": 1},
#    "MII225202": {"matakuliah": "Matematika untuk Ilmu Komputer", "hari": 4, "jam": 3, "ruangan": 1},
#    "MII226001": {"matakuliah": "Metodologi Riset Ilmu Komputer", "hari": 5, "jam": 3, "ruangan": 1}
#}Jadwal terbaik: [2, 3, 1, 11, 7, 6]
#Jadwal terbaik: [7, 11, 15, 10, 3, 8]
jadwal = {
    1 : {"MII225201": "Analisis Algoritme", "hari" :2,"jam" : 2,"ruangan": 1},
    2 : {"MII225202": "Matematika untuk Ilmu Komputer", "hari" :4,"jam" : 3,"ruangan": 1},
    3 : {"MII226001": "Metodologi Riset Ilmu Komputer", "hari" :5,"jam" : 3,"ruangan": 1},
    4 : {"MII225201": "Pengolahan dan Analisis Citra Digital", "hari" :4,"jam" : 3,"ruangan": 1},
    5 : {"MII226502": "Sistem Temu Balik Informasi", "hari" :4,"jam" : 1,"ruangan": 1},
    6 : {"MII226503": "Data Warehouse dan Inteligensi Bisnis", "hari" :2,"jam" : 1,"ruangan": 1},
    7 : {"MII226504": "Pengembangan Perangkat Lunak", "hari" :3,"sesi" : 3,"ruangan": 1},
    8 : {"MII226505": "Kecerdasan Digital dan Informatika Sosial", "hari" :1,"jam" : 3,"ruangan": 1},
    9 : {"MII226506": "Manajemen dan Audit Sistem Informasi", "hari" :1,"jam" : 3,"ruangan": 1},
    10 : {"MII226203": "Teori Komputasi", "hari" :4,"jam" : 2,"ruangan": 1},
    11 : {"MII226401": "Kecerdasan Komputasional dan Pembelajaran Mesin", "hari" :2,"jam" : 3,"ruangan": 1},
    12 : {"MII226402": "Rekayasa Fitur dan Pengenalan Pola", "hari" :2,"jam" : 2,"ruangan": 1},
    13 : {"MII226404": " Sistem Pendukung Pembuatan Keputusan ", "hari" :4,"sesi" : 2,"ruangan": 1},
    14 : {"MII226501": " Data Science ", "hari" :3,"sesi" : 2},
    15 : {" MII226601": " Komputasi Awan dan Keamanan Siber ", "hari" :2,"sesi" : 3,"ruangan": 1},
    16 : {" MII226602": " Jaringan Komputer Lanjut ", "hari" :3,"sesi" : 2,"ruangan": 1},
    17 : {" MII226603": " Platform dan Arsitektur Big Data ", "hari" :2,"sesi" : 2,"ruangan": 1},

}
# Parameter GA
pop_size = 10
generations = 50
mutation_rate = 0.1

# Fungsi fitness
def fitness(individual):
    penalty = 0
    for i in range(len(individual)):
        for j in range(i + 1, len(individual)):
            if individual[i] == individual[j]:  # Cek tabrakan jadwal dalam satu hari
                penalty += 1
            if abs(i // 3 - j // 3) == 0 and abs(individual[i] % 3 - individual[j] % 3) == 0:  # Cek tabrakan jadwal dalam satu sesi
                penalty += 1
    return 1 / (1 + penalty)

# Inisialisasi populasi awal
def generate_individual():
    individual = []
    for _ in range(6):
        #individual.append(random.randint(1, 20))
        individual.append(random.randint(1, 18))
    return individual

# Seleksi orang tua menggunakan roulette wheel
def select_parents(population):
    fitness_sum = sum([fitness(individual) for individual in population])
    rand = random.uniform(0, fitness_sum)
    current_sum = 0
    for individual in population:
        current_sum += fitness(individual)
        if current_sum >= rand:
            return individual

# Crossover menggunakan 1-point crossover
def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# Mutasi dengan pemilihan nilai acak
def mutate(individual):
    if random.random() < mutation_rate:
        index = random.randint(0, len(individual) - 1)
        individual[index] = random.randint(1, 60)
    return individual

# Algoritma Genetika
def genetic_algorithm():
    population = [generate_individual() for _ in range(pop_size)]
    for generation in range(generations):
        new_population = []
        for _ in range(pop_size // 2):
            parent1 = select_parents(population)
            parent2 = select_parents(population)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutate(child1)
            child2 = mutate(child2)
            new_population.extend([child1, child2])
        population = new_population
        best_individual = max(population, key=fitness)
        print("Generasi:", generation + 1, "Fitness:", fitness(best_individual))
        print("Jadwal terbaik:", best_individual)
        


# Fungsi untuk mencetak jadwal terbaik
def print_jadwal(individual):
    jadwal_terbaik = {}
    for i, idx in enumerate(individual):
        matakuliah_key = list(jadwal.keys())[i]
        matakuliah_value = list(jadwal.values())[i]
        jadwal_terbaik[matakuliah_key] = matakuliah_value
    print("Jadwal Terbaik:")
    for key, value in jadwal_terbaik.items():
        print(f"Matakuliah: {value}, Hari: {value['hari']}, Jam: {value['jam']}, Ruangan: {value['ruangan']}")

# Menjalankan algoritma genetika
genetic_algorithm()
````
